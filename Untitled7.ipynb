```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('house_price.csv')
data
```

```python
# Mean Function
```

```python
data['price_per_sqft'].mean()
```

```python
# Percentile Method
```

```python
percentiles = np.percentile(data['price_per_sqft'], [25, 50, 75])
print("Percentiles:",percentiles)
Minimum = data.price_per_sqft.quantile(0)
Minimum
```

```python

Maximum = data.price_per_sqft.quantile(0.99)
Maximum
```

```python
# IQR
```

```python
Q1 = data['price_per_sqft'].quantile(0.25)
Q3 = data['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR
print(lower_whisker)
print(upper_whisker)
```

```python
outliers= data[(data['price_per_sqft'] >= lower_whisker) & (data['price_per_sqft'] <= upper_whisker)]
print(outliers)
```

```python
#Normal Distribution
```

```python

sns.histplot(outliers['price_per_sqft'], kde=True)
plt.title('Histogram of Price per Sqft')
plt.xlabel('Price per Sqft')
plt.ylabel('Count')
plt.show()
```

```python
#ZScore Method
Z=(x-Âµ)/Ïƒ
```

```python
mean = data['price_per_sqft'].mean()
mean
```

```python
std = data['price_per_sqft'].std()
std
```

```python
z = (data['price_per_sqft']-mean)/std
print('Zscore is',z)
```

```python
data['Zscore']= z
data.head(10)
```
